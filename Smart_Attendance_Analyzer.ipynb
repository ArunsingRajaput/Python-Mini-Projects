```python
attendance_list = []

def count_present(total):
    present = 0
    for status in attendance_list:
        if status == "P" or status == "p":
            present += 1
    print("Present students:", present)
    return present

def attendance_percentage(present, total):
    percentage = (present / total) * 100
    print("Attendance percentage:", round(percentage, 2), "%\n")

print(" ~ Smart Attendance Analyzer ~ ")
total_students = int(input("Enter number of students : "))

for i in range(total_students):
    status = input("Enter attendance for student " + str(i+1) + " (P/A): ")
    attendance_list.append(status)

present_count = count_present(total_students)
attendance_percentage(present_count, total_students)
```
