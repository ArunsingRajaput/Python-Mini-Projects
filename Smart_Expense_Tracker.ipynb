```python
expenses = [] 

def add_expense():
    amount = float(input("Enter expense amount: "))
    category = input("Enter category (Items Name): ")
    record = {"amount": amount, "category": category}
    expenses.append(record)
    print("Expense added successfully!\n")

def view_expenses():
    if not expenses:
        print("No expenses recorded yet.\n")
        return
    print("---- All Expenses ----")
    for i, exp in enumerate(expenses, 1):
        print(i, "Amount:", exp["amount"], "| Category:", exp["category"])
    print()

def total_spending():
    total = 0
    for exp in expenses:
        total += exp["amount"]
    print("Total spending: ₹", total, "\n")

def category_analysis():
    if not expenses:
        print("No data available for analysis.\n")
        return
    analysis = {}
    for exp in expenses:
        cat = exp["category"]
        if cat in analysis:
            analysis[cat] += exp["amount"]
        else:
            analysis[cat] = exp["amount"]
    print("---- Category-wise Spending ----")
    for cat, total in analysis.items():
        print(cat, ": ₹", total)
    print()

while True:
    print("===== Smart Expense Tracker =====")
    print("1. Add Expense")
    print("2. View Expenses")
    print("3. Total Spending")
    print("4. Category-wise Analysis")
    print("5. Exit")
    ch = input("Choose an option: ")

    if ch == "1":
        add_expense()
    elif ch == "2":
        view_expenses()
    elif ch == "3":
        total_spending()
    elif ch == "4":
        category_analysis()
    elif ch == "5":
        print("Exiting... Stay smart with your expenses!")
        break
    else:
        print("Invalid option! Try again.\n")
```
