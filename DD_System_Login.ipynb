```python
def employee_Registration():
    employee_id = input("Enter Employee ID: ")
    name = input("Enter Name: ")
    age = int(input("Enter Age: "))
    department = input("Enter Department: ")
    basic_salary = float(input("Enter Basic Salary: "))

    if age >= 18 and basic_salary > 0:
        print("\nEmployee Registered Successfully!\n")
        return basic_salary
    else:
        print("\nError: Age must be ≥ 18 and Salary must be > 0\n")
        return None


def attendance():
    attend = []
    print("Enter attendance for 5 working days (P/A):")

    for i in range(5):
        at = input(f"Day {i+1}: ",)
        if at == "P" or at == "p":
            attend.append("P")
        else:
            attend.append("A")

    print("\nAttendance Record:", attend)
    present_count = attend.count("P")

    if present_count >= 4:
        print("Status: Regular\n")
    elif present_count == 3:
        print("Status: Warning\n")
    else:
        print("Status: Irregular\n")

    return present_count


def salary_calculation(bs, pd):
    hra = bs * 0.20
    da = bs * 0.10
    pf = bs * 0.12

    if pd < 3:
        net_salary = bs - pf
        print("Net Salary (No HRA & DA due to low attendance):", net_salary, "\n")
    else:
        net_salary = (bs + hra + da) - pf
        print("Net Salary:", net_salary, "\n")

    return net_salary


def performance():
    ratings = []
    print("Enter 3 Performance Ratings (1–5):")

    for i in range(3):
        r = int(input(f"Rating {i+1}: "))
        ratings.append(r)

    avg_rating = sum(ratings) / 3
    print("\nAverage Rating:", avg_rating)

    if avg_rating >= 4:
        print("Performance: Excellent\n")
    elif avg_rating >= 3:
        print("Performance: Good\n")
    else:
        print("Performance: Needs Improvement\n")

    return avg_rating

def login(username, password):
    if username == "Arun" and password == "Arun@123":
        print("\nLogin Successful! Welcome to DIGITAL DREAMS Admin Panel\n")
        return True
    else:
        print("\nAccess Denied! Invalid Credentials\n")
        return False


# MAIN PROGRAM 

print("===== DIGITAL DREAMS (DD) SYSTEM LOGIN =====")
user = input("Enter Username : ")
pwd = input("Enter Password : ")

if login(user, pwd):
    basic_salary = employee_Registration()

    if basic_salary is not None:
        present_days = attendance()
        salary_calculation(basic_salary, present_days)
        performance()

    print("Thank you for using DIGITAL DREAMS Employee System\n")
else:
    print("System Closed\n")
```
